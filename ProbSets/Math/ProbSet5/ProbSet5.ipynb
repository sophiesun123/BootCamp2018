```python
# EXERCISE 2
def intersection(a1, b1, c1, a2, b2, c2):
    det = a1 * b2 - a2 * b1
    if (det == 0):
        print("The lines are parallel.")

    else:
        x = (c1 * b2 - c2 * b1) / det
        y = (a1 * c2 - a2 * c1) / det
    return x, y
# i
x = np.linspace(0, 7, 1000)
x1 = (15 - x) / 3 
x2 = (18 - 2 * x) / 3 
x3 = x - 4 
plt.figure(figsize=(10, 10))
plt.plot(x, x1)
plt.plot(x, x2)
plt.plot(x, x3)
plt.axvline(x = 0)
plt.plot(x, np.zeros(1000))
plt.fill_between(x, x3, x1, color='blue', where = x2 >= x1, interpolate = True)
plt.fill_between(x, x3, x2, color='blue', where = x2 <= x1, interpolate = True)
plt.xlim(0, 7)
plt.ylim(0, 6)
o1 = 1 - 3 * x
o2 = 5 - 3 * x
o3 = 10 - 3 * x
o4 = 15 - 3 * x
plt.plot(x, o1)
plt.plot(x, o2)
plt.plot(x, o3)
plt.plot(x, o4)
plt.show()
print("The optimizer for this problem is at x = 6 and y = 2.")
# ii
x = np.linspace(0, 28, 1000)
y1 = (11 + x) #less than
y2 = (27 - x) #less than
y3 = (90 - 2 * x) / 5 #less than
plt.figure(figsize=(10, 10))
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.axvline(x = 0)
plt.plot(x, np.zeros(1000))
plt.fill_between(x, 0, y1, color='blue', where = y3 >= y1, interpolate = True)
plt.fill_between(x, 0, y2, color='blue', where = y2 <= y3, interpolate = True)
plt.fill_between(x, 0, y3, color='blue', where = (y3 <= y1) & (y3 <= y2), interpolate = True)
plt.xlim(0, 28)
plt.ylim(0, 17)
o1 = (1 - 4 * x) / 6
o2 = (50 - 4 * x) / 6
o3 = (100 - 4 * x) / 6
o4 = (125 - 4 * x) / 6
plt.plot(x, o1)
plt.plot(x, o2)
plt.plot(x, o3)
plt.plot(x, o4)
plt.show()
print(intersection(1, 1, 27, 2, 5, 90))
print("The optimizer for this problem is at x = 15 and y = 12.")
```

$Ax \preceq b $

$ A^Ty \preceq c $

maximize $ (-b)^Ty $

$ (-A)^Ty \preceq c $
